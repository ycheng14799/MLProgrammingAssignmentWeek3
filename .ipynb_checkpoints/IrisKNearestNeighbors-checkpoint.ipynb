```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets 
from sklearn import neighbors
from sklearn.model_selection import validation_curve
from sklearn.decomposition import PCA
from matplotlib.colors import ListedColormap
from sklearn.preprocessing import scale
from sklearn.model_selection import ShuffleSplit
from matplotlib.patches import Patch
```

```python
# load iris dataset 
iris = datasets.load_iris()
# Inputs 
X = iris.data
y = iris.target
# Scale inputs 
scaledX = scale(X)
```

```python
# Validation curve graphing
def generateGraph(variable, trainScoresMean, testScoresMean,\
    trainScoresStd, testScoresStd,\
    xLabel, yLabel, title, log=False):
    plt.figure()
    if log is False:
        plt.plot(variable, trainScoresMean, "-o", color="blue",\
            label="Training Accuracy")
        plt.plot(variable, testScoresMean, "-o", color="red",\
            label="Cross-Validation Accuracy")
    else: 
        plt.semilogx(variable, trainScoresMean, "-o",\
            color="blue", label="Training Accuracy")
        plt.semilogx(variable, testScoresMean, "-o",\
            color="red", label="Cross-Validation Accuracy")
    plt.fill_between(variable, trainScoresMean-trainScoresStd,\
        trainScoresMean+trainScoresStd, alpha=0.1, color="blue")
    plt.fill_between(variable, testScoresMean-testScoresStd,\
        testScoresMean+testScoresStd, alpha=0.1, color="red")
    plt.xlabel(xLabel)
    plt.ylabel(yLabel)
    plt.title(title)
    plt.legend(loc="lower right")
```

```python
# Let us firstly evaluate the effect of K on the model prediction accuracy 
NNBasic = neighbors.KNeighborsClassifier()
# Ks to evaluate
Ks = np.arange(1,100,5,dtype=int)
# Random permutation cross-validator
cv = ShuffleSplit(n_splits=10, test_size=0.20)
# Generate validation curve
trainScores, testScores = validation_curve(NNBasic, scaledX, y, 'n_neighbors', Ks,\
    cv=cv, scoring="accuracy")
# Calculate mean and std 
trainScoresMean = np.mean(trainScores, axis=1)
testScoresMean = np.mean(testScores, axis=1)
trainScoresStd = np.std(trainScores, axis=1)
testScoresStd = np.std(testScores, axis=1)
# Graph
generateGraph(Ks, trainScoresMean, testScoresMean,\
    trainScoresStd, testScoresStd, "K", "Classification Accuracy", "Validation Curve for K (KNN)", log=False)
```

```python
# Let us try to visualize the decision boundaries in 2D 
# Idea: Projection into 2D with PCA 
# Classify with 4D features, then plot in 2D 
pca = PCA(n_components=2)
pca.fit(scaledX)
X2D = pca.transform(scaledX)

# Use optimal K value determined from previous step (approx: 11)
NNBasic = neighbors.KNeighborsClassifier(n_neighbors=11)
NNBasic.fit(scaledX, y)

# Define plot range 
plotXMin, plotXMax, plotYMin, plotYMax =\
    X2D[:,0].min() - 1,\
    X2D[:,0].max() + 1,\
    X2D[:,1].min() - 1,\
    X2D[:,1].max() + 1

# Define plot step-size
step = 0.01
plotXs, plotYs = np.meshgrid(np.arange(plotXMin, plotXMax, step),\
                            np.arange(plotYMin, plotYMax, step))

# Transform back to original space 
predictions = NNBasic.predict(pca.inverse_transform(np.c_[plotXs.ravel(), plotYs.ravel()]))
# Perform predictions in original 4D feature space
predictions = predictions.reshape(plotXs.shape)

# Define color maps for plot 
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

# Plot decision boundaries
plt.figure()
plt.pcolormesh(plotXs, plotYs, predictions, cmap=cmap_light)

# Plot training points 
plt.scatter(X2D[:,0], X2D[:,1], c=Y, cmap=cmap_bold,\
            edgecolor='k', s=20)
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.suptitle("Visualization of KNN Decision Boundaries (2D Projection using PCA)")
labels = ['Setosa', 'Versicolour', 'Virginica']
legendElements = [\
    Patch(facecolor='#FF0000', edgecolor='#FF0000',\
         label=labels[0]),\
    Patch(facecolor='#00FF00', edgecolor='#00FF00',\
         label=labels[1]),
    Patch(facecolor='#0000FF', edgecolor='#0000FF',\
         label=labels[2]),
]
plt.legend(handles=legendElements)
```

```python
# Alternative visualization
# Decision boundaries for a KNN model fit using a subset of 2 features 
# Select K=10
def plotDecisionBoundary2Features(f1,f2,xLabel,yLabel):
    X2F = scaledX[:,(f1,f2)]
    NNBasic = neighbors.KNeighborsClassifier(n_neighbors=10)
    NNBasic.fit(X2F, y)

    # Define plot range 
    plotXMin, plotXMax, plotYMin, plotYMax =\
        X2F[:,0].min() - 1,\
        X2F[:,0].max() + 1,\
        X2F[:,1].min() - 1,\
        X2F[:,1].max() + 1

    # Define plot step-size
    step = 0.01
    plotXs, plotYs = np.meshgrid(np.arange(plotXMin, plotXMax, step),\
                                np.arange(plotYMin, plotYMax, step))

    # Transform back to original space 
    predictions = NNBasic.predict(np.c_[plotXs.ravel(), plotYs.ravel()])
    # Perform predictions in original 4D feature space
    predictions = predictions.reshape(plotXs.shape)

    # Define color maps for plot 
    cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
    cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

    # Plot decision boundaries
    plt.figure()
    plt.pcolormesh(plotXs, plotYs, predictions, cmap=cmap_light)

    # Plot training points 
    plt.scatter(X2F[:,0], X2F[:,1], c=Y, cmap=cmap_bold,\
                edgecolor='k', s=20)
    plt.xlabel(xLabel)
    plt.ylabel(yLabel)
    plt.suptitle("Visualization of KNN Decision Boundaries (" + xLabel + ", " + yLabel +", K=10)")
    labels = ['Setosa', 'Versicolour', 'Virginica']
    legendElements = [\
        Patch(facecolor='#FF0000', edgecolor='#FF0000',\
             label=labels[0]),\
        Patch(facecolor='#00FF00', edgecolor='#00FF00',\
             label=labels[1]),\
        Patch(facecolor='#0000FF', edgecolor='#0000FF',\
             label=labels[2]),\
    ]
    plt.legend(handles=legendElements)
fNames = ["Sepal Length", "Sepal Width", "Petal Length", "Petal Width"]
for i in range(0,3):
    for j in range(i+1,4):
        plotDecisionBoundary2Features(i,j,fNames[i], fNames[j])
```

```python
# For the purposes of demonstrating the effects of K, 
# it is sufficient just to fit the KNN model to a subset of 2 features as before
# and vary K. We will assume that this behaviour generalizes to greater than 
# 2 dimensions and also to other features. 
# Let us fit a KNN model on the sepal width and sepal length features of the dataset

def plotVaryK(f1,f2,xLabel,yLabel,k):
    X2F = scaledX[:,(f1,f2)]
    NNBasic = neighbors.KNeighborsClassifier(n_neighbors=k)
    NNBasic.fit(X2F, y)

    # Define plot range 
    plotXMin, plotXMax, plotYMin, plotYMax =\
        X2F[:,0].min() - 1,\
        X2F[:,0].max() + 1,\
        X2F[:,1].min() - 1,\
        X2F[:,1].max() + 1

    # Define plot step-size
    step = 0.01
    plotXs, plotYs = np.meshgrid(np.arange(plotXMin, plotXMax, step),\
                                np.arange(plotYMin, plotYMax, step))

    # Transform back to original space 
    predictions = NNBasic.predict(np.c_[plotXs.ravel(), plotYs.ravel()])
    # Perform predictions in original 4D feature space
    predictions = predictions.reshape(plotXs.shape)

    # Define color maps for plot 
    cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
    cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

    # Plot decision boundaries
    plt.figure()
    plt.pcolormesh(plotXs, plotYs, predictions, cmap=cmap_light)

    # Plot training points 
    plt.scatter(X2F[:,0], X2F[:,1], c=Y, cmap=cmap_bold,\
                edgecolor='k', s=20)
    plt.xlabel(xLabel)
    plt.ylabel(yLabel)
    plt.suptitle("Visualization of KNN Decision Boundaries (" + xLabel + ", " + yLabel +", K=" + str(k) + ")")
    labels = ['Setosa', 'Versicolour', 'Virginica']
    legendElements = [\
        Patch(facecolor='#FF0000', edgecolor='#FF0000',\
             label=labels[0]),\
        Patch(facecolor='#00FF00', edgecolor='#00FF00',\
             label=labels[1]),\
        Patch(facecolor='#0000FF', edgecolor='#0000FF',\
             label=labels[2]),\
    ]
    plt.legend(handles=legendElements)

Ks = [1, 5, 10, 25, 50, 100]
for K in Ks: 
    plotVaryK(0,1,fNames[0],fNames[1],K)
```
